```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from numpy import median, mean
sns.set_theme()
# sns.set_theme(style="ticks")
```

# Load data

```python
tips = sns.load_dataset("tips")
tips['Tip [%]'] = (tips['tip']/tips['total_bill'])*100
tips
```

```python
tips.groupby(['time']).agg(
    count       = ('Tip [%]', 'count'), 
    mean        = ('Tip [%]', 'mean'),
    min         = ('Tip [%]', 'min'),
    perc_05     = ('Tip [%]', lambda x: x.quantile(0.05)),
    quantile_Q1 = ('Tip [%]', lambda x: x.quantile(0.25)),
    median      = ('Tip [%]', 'median'),
    quantile_Q3 = ('Tip [%]', lambda x: x.quantile(0.75)),
    perc_95     = ('Tip [%]', lambda x: x.quantile(0.95)),
    perc_99     = ('Tip [%]', lambda x: x.quantile(0.99)),
    max         = ('Tip [%]', 'max'),
)
```

# Plot comparison

```python
x = "Tip [%]"
y = "time"
plot_args = {'x':x, 'y':y, 'data':tips}
```

```python
f, ax = plt.subplots(figsize=(12, 5))
df_plot = tips.groupby(y)[x].mean().reset_index()
sns.barplot(x=x, y=y, data=df_plot)

# Tweak the visual presentation
ax.set_title('Pre-calculated mean')
ax.xaxis.grid(True)
ax.set(ylabel="");
ax.set_xlim(0,17);
```

```python
f, ax = plt.subplots(figsize=(12, 5))
df_plot = tips.groupby(y)[x].median().reset_index()
sns.barplot(x=x, y=y, data=df_plot)

# Tweak the visual presentation
ax.set_title('Pre-calculated median')
ax.xaxis.grid(True)
ax.set(ylabel="");
ax.set_xlim(0,17);
```

```python
f, ax = plt.subplots(figsize=(12, 5))
sns.barplot(**plot_args) # Default estimator is mean

# Tweak the visual presentation
ax.set_title('Seaborn barplot estimator default (mean)')
ax.xaxis.grid(True)
ax.set(ylabel="");
ax.set_xlim(0,18);
```

```python
f, ax = plt.subplots(figsize=(12, 5))
sns.barplot(**plot_args, estimator=median)

# Tweak the visual presentation
ax.set_title('Seaborn barplot estimator set to median')
ax.xaxis.grid(True)
ax.set(ylabel="");
ax.set_xlim(0,18);
```

# Scatterplots

```python
# Primary line
num1 = 10000
linear_x1 = np.random.uniform(0,100, num1) 
noise1 = np.random.normal(0, 90, num1)
linear_y1 = 5*linear_x1 + 25+noise1

# Hidden line
num2 = 3000
linear_x2 = np.random.uniform(40,60, num2) 
noise2 = np.random.normal(25, 28, num2)
linear_y2 = -10*linear_x2 + 750+noise2

# Gather the data and make a dataframe
data_x = np.hstack((linear_x1, linear_x2))
data_y = np.hstack((linear_y1, linear_y2))
df = pd.DataFrame.from_dict({'x':data_x, 'y':data_y})

plot_args = {'x':'x', 'y':'y', 'data':df}

print(len(df))
df.head()
```

```python
f, ax = plt.subplots(figsize=(9, 9))
sns.scatterplot(**plot_args, estimator=median)

# Tweak the visual presentation
ax.set_title('Seaborn scatterplot')
ax.xaxis.grid(True)
ax.set(ylabel="");
ax.set_xlim(-5, 105);
ax.set_ylim(-300, 800);
```

```python
f, ax = plt.subplots(figsize=(9, 9))
sns.histplot(**plot_args, bins=50, pthresh=.1) # pthresh=.1
sns.kdeplot(**plot_args, levels=10, thresh=.1) # thresh=.1

# Tweak the visual presentation
ax.set_title('Seaborn kdeplot and histplot highlighting hidden structure')
ax.xaxis.grid(True)
ax.set(ylabel="");
ax.set_xlim(-5, 105);
ax.set_ylim(-300, 800);
```
